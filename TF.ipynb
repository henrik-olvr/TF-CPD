```python
import pandas as pd
tabela = pd.read_csv("dados/MICRODADOS_ENEM_2000.csv", sep=";")

display(tabela)
```

```python
municipios = tabela[["NO_MUNICIPIO_RESIDENCIA"]].groupby("NO_MUNICIPIO_RESIDENCIA").sum()
display(municipios)
```

```python
# PESQUISANDO UM MUNICIPIO
nome_municipio = "PORTO ALEGRE"
municipio = tabela[tabela.NO_MUNICIPIO_RESIDENCIA==nome_municipio].count()[0]
print("NÃºmero de apariÃ§Ãµes do municipio '{}' = {}".format(nome_municipio, municipio))
tabela_municipio = tabela[tabela.NO_MUNICIPIO_RESIDENCIA==nome_municipio]
display(tabela_municipio)

# CHECANDO CANDIDATOS QUE ESTIVERAM PRESENTES E AUSENTES NA PROVA
canditatos_presentes = tabela_municipio[tabela_municipio.TP_PRESENCA==1]
canditatos_ausentes = tabela_municipio[tabela_municipio.TP_PRESENCA==0]
display(canditatos_presentes)
display(canditatos_ausentes)

# OBTENDO NUMERO DE INSCRICAO DOS CANDIDATOS AUSENTES NA PROVA
display(canditatos_ausentes.NU_INSCRICAO)
```
